```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate 1000 random numbers uniformly between [0, 1]
numbers = np.random.uniform(0, 1, 1000)

# Plot histogram with 100 bins
plt.hist(numbers, bins=100, edgecolor='black')

# Add labels 
plt.xlabel('Numbers')
plt.ylabel('Frequency')

# Save figure as a PDF
plt.savefig('Session9.pdf')

# Display plot
plt.show()
```
