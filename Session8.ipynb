```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Create array from [0, 1] in 100 steps
x = np.linspace(0, 1, 100)

# Define functions sin(x) and cos(x)
def sine(x):
    return np.sin(x)

def cosine(x):
    return np.cos(x)
```

```python
# Create multipanel plot with 1 row and 2 columns
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 8))

# Plot sin(x) in the left
axes[0].plot(x, sine(x))
axes[0].set_title('sin(x)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')

# Plot cos(x) in the right
axes[1].plot(x, cosine(x))
axes[1].set_title('cos(x)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('cos(x)')

# Save the figure as a PDF
plt.savefig('Session8.pdf')

# Displaying the plot
plt.show()
```
