```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(30)
#Generate 1000 random numbers exponential
ex_numbers = np.random.exponential(scale=1, size=1000)

# Plot  histogram
plt.hist(ex_numbers, bins=100, density=True, alpha=0.75, color='purple', edgecolor='black')

# Add labels
plt.xlabel('Number')
plt.ylabel('Frequency')

# Save as PDF
plt.savefig('Session10.pdf')

# Display the plot
plt.show()
```
